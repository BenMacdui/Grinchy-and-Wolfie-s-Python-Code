```python
#@title Word guessing Game
import random
from IPython.display import clear_output
import time

animals = ['ants','bear','snake', 'wolf', 'tiger', 'elephant', 'wale', 'spider', 'deer', 'owl', 'eagle', 'seal', 'polar bear', 'lion', 'monkey', 'aardvark', 'kangaroo', 'mouse', 'rat', 'butterfly', 'wasp', 'bee', 'fox', 'badger' ]
towns = ['paris','london', 'berlin', 'southport', 'stockport', 'glasgow', 'edinburgh', 'aviemore', 'hamburg', 'amsterdam', 'manchester', 'liverpool']
nature = ['beach', 'sky', 'tree', 'flower', 'moutains', 'loch', 'sea', 'ocean', 'clouds', 'stars', 'rocks', 'beach', 'meadow', 'sandbank', 'hay field', 'conifer']
sport = ['football', 'netball', 'hockey', 'ice skating', 'rock climbing', 'netball', 'basketball', 'taekwondo', 'bowling', 'darts', 'handball']
categories = ['animals', 'towns', 'nature', 'sport']
b = random.choice(categories)
x = "animals"
y = "towns"
z = 'nature'
v = 'sport'

if x in b:
    cat = animals
    print("the categorie is animals")
elif y in b:
    cat = towns
    print("the categorie is towns")
elif z in b:
    cat = nature
    print("the category is nature")
else:
    cat = sport
    print("the category is sport")

chosen_word = random.choice(cat)
print()
print("The word you have to guess has", len(chosen_word), "characters")
print()


response = ['Nope', "Wrong again", 'Need to get your brain working, this is not right', 'You missed it', 'better luck next time', 'Wolfie would do better. Try again', 'Oh, no need to cry, you will get it right eventually', 'I can hear plankton laughing']

display = []
word_length = len(chosen_word)
for _ in range(word_length):
    display += "_"
     

def give_letter():
    if cat == animals:
        print("the category is animals")
        print()
    elif cat == towns:
        print("the category is towns")
    elif cat == nature:
        print("the category is nature")
    else:
        print("the category is sport")
                

    position = 0
    run = 0
    i = 0
    print(chosen_word)
    print()
    tries = input(f"I will choose a letter for you\nYou have {word_length - 2} tries. \nSometimes I choose a letter twice \nTough cookie \nIt still counts as a try \nThe word has {word_length} characters \nType enter to get help or the word  ").lower()
    if tries == chosen_word:
        print()
        print("You were lucky this time. Correct. How lucky can you be? ")
        print()
    elif tries =="":
        for i in range(word_length - 2):
            b = random.sample(range(word_length),word_length)
            c = b[position]
            y = chosen_word[c]
            if y in display:
                print('already guessed the letter. Tough')
                # print(random.choice(response))
                # break
            else:
                display[c] = y
                print(display)
                run += 1
                position += 1
            i += 1
            tries = input(">>> enter for more help or word  ").lower()
            if tries == chosen_word:
                  print("You were lucky this time. Correct. Get you next time")
                  break
            elif i == word_length -2:
              print()
              print(random.choice(response))
              print()
    elif tries != chosen_word and tries != "":
        print()
        print(random.choice(response))
                
        
    print()
        
       
        
    return

print(f"""the category list for a quick overview of words:
        {cat}""")
time.sleep(5)
clear_output()

print()
print()
print()

guess = input("Guess the word or press enter for help ").lower()
for w in chosen_word:
    y = w
if guess == chosen_word:
    print('You right')
elif guess == "":
    give_letter()
else:
    print(random.choice(response))
    
    

print()



```
