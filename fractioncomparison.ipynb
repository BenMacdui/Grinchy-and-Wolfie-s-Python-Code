```python
"""This part of the fraction progamme written by Wolfie, the trainee pirate, takes as its input
two fractions to compare. It could do it pretty fast but Wolfie wants us to see how
fractions are compared by hand so to speak. So we go through the steps with the program together. 
"""


from fractions import Fraction as frac
import time
import sys
import math
from prettytable import PrettyTable

a, b = input('Enter the two fractions to be compared to, in the form of a/b to, sperated by a space: ').split()
print()
x = frac(a)   
y = frac(b)
m = 15 # defines the range of multipliers

d1 = x.denominator
d2 = y.denominator
m1 = list(range(d1, (m+1)*d1, d1))
m2 = list(range(d2, (m+1)*d2, d2))

example = PrettyTable([' ', 4, 6])
example.add_row(['1','4','6'])
example.add_row(['2','8','12'])
example.add_row(['3','12','18'])
print()
print(f"""When you compare two fractions you need to find the lowest common denominator
in order to do so you would write a table with the multipliers of the two denominators
This program replicates this. The left hand side of the table is a count column. 


It tells us by how much we have to multiply each fraction by. 

The other columns are the multipliers of the denominators. The lowest common denominator can be found in 
these columns. For each list we read off the number on the left hand side of the table and then multiply 
our fraction by that number.

For example if we have a fraction 2/4 and 5/6 and the lowest commond denominator is 12, then we find 
in the column for 4 (one of the denominators) the 12 in 3rd place so we multiply 2/4 * 3 and 
we find the 12 in the column under 6 (the second deonominator) at 2nd place, 
so we multiply the fraction 5/6 * 2


{example}


I hope this is relatively clear. 

Let's have a look at the table for the fractions you choose to compare""")

time.sleep(3)
print()
print('\U0001f600')
print()
cont = input('hit enter to continue')
if cont == " ":
    print()
    print()
print()
print()



#we set up the table row names
# we insert the variables here so everytime we have new fractions
# we have different table row names
t = PrettyTable([' ', d1, d2])

"""the next bid sets a counter than iterates over two lists. 
zip stops when the shorter (if there is a shorter) list stops.
Every iteration we add a row to our table created at the top.
Once the loop is finished we print the table (print(t))"""
count = 0
for items, items1 in zip(m1, m2):
    count += 1
    t.add_row([count, items, items1])
print(t)    
print()
print()

def lcm(d1, d2):
    import math 
    
    return  d1 * d2 // math.gcd(d1, d2)

lcm(d1,d2)
print(f"The lowest common denominator for your chosen fractions = {lcm(d1,d2)}")
time.sleep(3)
print('Have a stab at what the multipliers are before you continue \U0001f600')
enter = input('hit enter to see the multiplier')
if enter == " ":
    print()
    print()
# returns the count (multiplier) for each list
# don't forget to add 1 to each index counting in python starts at 0
index1 = m1.index(lcm(d1,d2))
index2 = m2.index(lcm(d1,d2))
print()
print(f"the multiplier for {x} is {index1 + 1}\nthe multiplier for {y} is {index2 + 1}")
print()
print("Did you get this right?")
time.sleep(4)
print('now what is the new fraction?')
print()
time.sleep(3)
enter1 = input("do you want to know? Do you have the answer? Hit enter to find out \U0001f600")
if enter1 == " ":
    print()
    print()
print()
i1 = index1+1
i2 = index2+1
h1 = str(f"{x.numerator*i1}/{x.denominator * i1}")
h2 = str(f"{y.numerator*i2}/{y.denominator * i2}")

print('The fractions are:')
print()
print(h1, end = " and ")
print(h2, end = "")
print()
print()
print('Now we can easily compare them and see which one is the greater one')
print()
print("Good bye from Wolfie the trainee pirate \U0001f600")

```
