```python
# this simple programme stems from https://thehelloworldprogram.com/python/python-game-rock-paper-scissors/
# We are making ammendments to it and changing things for learning purposes


#@title Rock Paper and Sissors

from random import randint
import time
from IPython.display import clear_output


# create a list of play options
t = ['rock', 'sissors', 'paper']

#assign the computer random play
computer = t[randint(0,2)]


# We set our counters to zero and then add into them to keep
# a tally on the score
computercount = c = 0
playercount = p = 0
tie = 0


# set run to true so that it runs while true, till interrupted with q
run = True

# the main loop: this will run till the user inputs q (for quit)
while run:
    
    
#after the beginning of the main loop (here a while loop),
# we clear all output from before. It makes for a nicer screen

    clear_output()


# turn all input into lower case alphabet with .lower() function, 
#just in case a user has all capitals on. 

    player = input('rock, sissors, paper? or q to quit ').lower()

    
#after our input we get our conditionals. This is to evaluate the options
# what did the computer choose, what did the player choose, who wins.

    if player == computer:
        print('computer chose', computer, )
        print("tie")
        tie += 1
        print(f"Score: You {tie} : Computer {tie}")
        time.sleep(3)

    elif player == 'rock':
        if computer == 'paper':
            print('computer chose', computer, )
            print('\nyou loose', computer, 'covers', player)
            c += 1
            print(f"Score: You {p} : Computer {c}")
            time.sleep(3)
        else: 
            print('computer chose', computer, )
            print('You win.', player, 'smashes', computer)
            p += 1
            print(f"Score: You {p} : Computer {c}")
            time.sleep(3)

    elif player == 'paper':
        if computer == 'sissors':
            print('computer chose', computer,)
            print('\nyou loose', computer, 'cuts', player)
            c  += 1
            print(f"Score: You {p} : Computer {c}")
            time.sleep(3)
        else:
            print('computer chose', computer, )
            print('you win', player, 'covers', computer)
            p += 1
            print(f"Score: You {p} : Computer {c}")
            time.sleep(3)

    elif player == 'sissors':
        if computer == 'rock':
            print('computer chose', computer,)
            print('\nyou loose', computer, 'smashes', player)
            c += 1
            print(f"Score: You {p} : Computer {c}")
            time.sleep(3)
        else:
            print('computer chose', computer, )
            print('you win', player, 'cut', computer)
            p += 1
            print(f"Score: You {p} : Computer {c}")
            time.sleep(3)

# If you typed in q the computer will break the loop and end the game
    elif player == 'q':
      print("Thanks for playing")
      break
# if the user accidentially made a typo we tell the user then go back to the 
# input after 2 seconds    
    else:
        print('That is not a valid choice. Check your spelling')
    time.sleep(2)
    

# Once we broken out of the loop we display the final score 
print()
print(f"Total score: You {p} vs Computer {c}. \nthere were {tie} ties")
```
